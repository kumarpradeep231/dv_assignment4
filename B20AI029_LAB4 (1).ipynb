1.1 **Monthly Changes In Different Food Grain Stocks**

**HEATMAPS**

```python
# Set up the figure with 4 subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))
fig.suptitle('Monthly Changes in Different Food Grain Stocks', fontsize=16)

# Set up the color palettes
cmap_continuous = 'YlGnBu'
cmap_onecolor = 'YlOrBr'
cmap_bipolar = sns.color_palette("coolwarm", as_cmap=True)
cmap_custom = ['#FEE08B', '#FDAE61', '#F46D43', '#D53E4F', '#9E0142']

# Continuous color palette
axs[0, 0].set_title('Continuous Color Palette')
sns.heatmap(df.iloc[:,3:], cmap=cmap_continuous, ax=axs[0, 0], cbar_kws={'label': 'Stock Levels (Metric Tons)'})

# One color sequential palette
axs[0, 1].set_title('One Color Sequential Palette')
sns.heatmap(df.iloc[:,3:], cmap=cmap_onecolor, ax=axs[0, 1], cbar_kws={'label': 'Stock Levels (Metric Tons)'})

# Bipolar sequential palette
axs[1, 0].set_title('Bipolar Sequential Palette')
sns.heatmap(df.iloc[:,3:], cmap=cmap_bipolar, center=0, ax=axs[1, 0], cbar_kws={'label': 'Stock Levels (Metric Tons)'})

# Custom color palette
axs[1, 1].set_title('Custom Color Palette')
sns.heatmap(df.iloc[:,3:], cmap=cmap_custom, ax=axs[1, 1], cbar_kws={'label': 'Stock Levels (Metric Tons)'})

plt.show()
```

**Line Plots**

```python
# Define the custom color palette
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

# Create the subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Continuous color palette
axs[0, 0].set_title('Continuous Color Palette')
for index, row in df.iterrows():
    if row['Particulars'].startswith('Foodgrains '):
        label = row['Particulars'][11:]
        data = row[3:]
        axs[0, 0].plot(data, label=label, color=colors[index])
axs[0, 0].set_xlabel('Month')
axs[0, 0].set_ylabel('Stock Levels (Metric Tons)')
axs[0, 0].legend()

# One color sequential palette
axs[0, 1].set_title('One Color Sequential Palette')
for index, row in df.iterrows():
    if row['Particulars'].startswith('Foodgrains '):
        label = row['Particulars'][11:]
        data = row[3:]
        color = plt.get_cmap('Blues')(index/10)
        axs[0, 1].plot(data, label=label, color=color)
axs[0, 1].set_xlabel('Month')
axs[0, 1].set_ylabel('Stock Levels (Metric Tons)')
axs[0, 1].legend()

# Bipolar sequential palette
axs[1, 0].set_title('Bipolar Sequential Palette')
for index, row in df.iterrows():
    if row['Particulars'].startswith('Foodgrains '):
        label = row['Particulars'][11:]
        data = row[3:]
        color = plt.get_cmap('coolwarm')(index/10)
        axs[1, 0].plot(data, label=label, color=color)
axs[1, 0].set_xlabel('Month')
axs[1, 0].set_ylabel('Stock Levels (Metric Tons)')
axs[1, 0].legend()

# Custom color palette
axs[1, 1].set_title('Custom Color Palette')
for index, row in df.iterrows():
    if row['Particulars'].startswith('Foodgrains '):
        label = row['Particulars'][11:]
        data = row[3:]
        color = colors[index]
        axs[1, 1].plot(data, label=label, color=color)
axs[1, 1].set_xlabel('Month')
axs[1, 1].set_ylabel('Stock Levels (Metric Tons)')
axs[1, 1].legend()

# Adjust the layout and show the plots
plt.tight_layout()
plt.show()
```

#Part -2 State wise Colleges of different Ayus Category

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
df = pd.read_csv("/content/MergerdAyush_Colleges_in_India_01-04-2021_0 (3).csv")
df.head()
```

**Grouping The Collage Corresponding To Their State And Type Of System**

```python
# Group the data by State and Ayus category, and sum the columns for number of colleges
df1 = df.groupby(['State or Union Territory', 'Type of System'])['No. of Colleges - Govt', 'No. of Colleges - Non-Govt'].sum().reset_index()
```

```python
df1.head()
```

**Making Catagories of Different Type Of System and GOV vs NON - GOV**

```python
# Pivot the data to make the Ayus categories columns and states rows
df2 = df1.pivot(index='State or Union Territory', columns='Type of System', values=['No. of Colleges - Govt', 'No. of Colleges - Non-Govt'])
df2.head()
```

**Stacked Barplot**

```python

fig, ax = plt.subplots(figsize=(10, 8))
plt.title('State-wise Colleges of different Ayus Categories')
plt.ylabel('Number of Colleges')
plt.legend(title='Type of System')
df2.plot(kind='bar', stacked=True, ax=ax)
```

**Grouped Barplot**

```python

figsize = (10, 6)


fig, ax = plt.subplots(figsize=figsize)
df2.plot(kind='bar', width=1.5, ax=ax)


plt.title('State-wise Colleges of different Ayus Categories')
plt.xlabel('State or Union Territory')
plt.ylabel('Number of Colleges')
plt.legend(title='Type of System')
```

# 3.Distribution of National, State and District roads.

```python
# Load the CSV file
df = pd.read_csv('/content/Road_Transport_BRS_2017-18_Annexure7_9c (1).csv')
df.head()
```

```python

national_highways = df['National Highways']
state_highways = df['State Highways']
district_roads = df['District Roads']

# Plot the distributions
plt.hist(national_highways, bins=20, alpha=0.5, label='National Highways')
plt.hist(state_highways, bins=20, alpha=0.5, label='State Highways')
plt.hist(district_roads, bins=20, alpha=0.5, label='District Roads')

# Add axis labels and legend
plt.xlabel('Road Length')
plt.ylabel('Frequency')
plt.legend(loc='upper right')

# Show the plot
plt.show()
```
